```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("pjsk.csv")
df.columns
```

```python
df.head(26)
```

```python
df.at[1, 'Class Year']='None'
df.at[2, 'Class Year']='None'
df.at[3, 'Class Year']='None'
df.at[4, 'Class Year']='None'
df.at[5, 'Class Year']='None'
df.at[0, 'Class Year']='None'
df.at[22, 'Class Year']='None'
df.head(26)
```

```python
classYear = {"None": 0, "Second Year": 0, "Third Year": 0}
for character in df['Class Year']:
    if character == 'None':
        classYear['None'] += 1
    elif character == 'Second Year':
        classYear['Second Year'] += 1
    elif character == 'Third Year':
        classYear['Third Year'] += 1

freqTable = {
    "Classes": ["None", "Second Year", "Third Year"],
    "Amount": [classYear["None"], 
               classYear["Second Year"],
               classYear["Third Year"]]
}

freqTable = pd.DataFrame(freqTable)
freqTable.head()
```

```python
sns.barplot(data=freqTable, x="Classes", y="Amount")
```

```python
df_classone = df[df["Classes"] == "None"]
sns.histplot(data=df_classo, x="Classes", bins=100, color="#0077DD")
```
