```python
import pandas as pd
import seaborn as sns

em = pd.read_json("english_movies.json")
em.info()
```

```python
fm = pd.read_json("french_movies.json")
fm.info()
```

```python
em.head()
```

```python
fm.head()
```

```python
sns.regplot(data=em, x="budget", y="popularity")
sns.regplot(data=fm, x="budget", y="popularity")
```

```python
import numpy as np
em["popularity_sqrt"] = np.sqrt(em["popularity"])
em["budget_sqrt"] = np.sqrt(em["budget"])
em.head()
```

```python
fm["popularity_sqrt"] = np.sqrt(fm["popularity"])
fm["budget_sqrt"] = np.sqrt(fm["budget"])
fm.head()
```

```python
sns.regplot(data=em, x="budget_sqrt", y="popularity_sqrt")
sns.regplot(data=fm, x="budget_sqrt", y="popularity_sqrt")
```

```python
em['popularity'].max()
```

```python
threshold = 6191
em_filtered = em[em['popularity'] < threshold]
```

```python
sns.regplot(data=em_filtered, x="budget", y="popularity")
sns.regplot(data=fm, x="budget", y="popularity")
```

```python
em['popularity_sqrt'].max()
```

```python
tree = 78.68601527590528
em_filtered = em[em['popularity_sqrt'] < tree]
```

```python
sns.regplot(data=em_filtered, x="budget_sqrt", y="popularity_sqrt")
sns.regplot(data=fm, x="budget_sqrt", y="popularity_sqrt")
```

```python
fm['popularity_sqrt'].max()
```

```python
yahoo = 13.047490180107436
fm_filtered = fm[fm['popularity_sqrt'] < yahoo]
```

```python
sns.regplot(data=em_filtered, x="budget_sqrt", y="popularity_sqrt")
sns.regplot(data=fm_filtered, x="budget_sqrt", y="popularity_sqrt")
```

<span style="color: purple; font-size: 15px;">The regresion line for English movies, blue, have a steeper slope than the French movies, orange,  showing that the budget for English movies have a signnificantly higher impact on the popularity of movies in English than French. A lot of French movies have a low budget, which give them a lower popularity score compared to the one movie with a higher buget, but still lower popularity score than most of the English movies with around the same budget.</span>
