# My Question

<h3 style="text-align: center; color: #3366cc;">What is the probability of randomly selecting a game from 2020 and after?</h3>

# My Analysis

```python
import pandas as pd
import requests
import time

api_key = '416b4b821ed842ce8f81f836f812d269'
endpoint = 'https://api.rawg.io/api/games'

params = {
    'key': api_key,
    'page_size': 50,
    'page': 1
}

all_games = []

for page in range(1, 51):
    params['page'] = page
    response = requests.get(endpoint, params=params)

    if response.status_code == 200:
        gd = response.json()
        games = gd['results']
        
        if not games:
            print(f"No games found on page {page}. Stopping.")
            break
        
        all_games.extend(games)
        print(f"Fetched page {page} with {len(games)} games")
        
        time.sleep(1)
    else:
        print(f"Error fetching page {page}: {response.status_code}")
        break

df = pd.DataFrame(all_games)

print(f"Total games fetched: {len(df)}")
print(df.head())
```

```python
df
```

```python
import seaborn as sns
import random

df['released'] = pd.to_datetime(df['released'])

df['release_year'] = df['released'].dt.year

sns.histplot(data=df, x="released")
```

<p style="text-align: center; color: #ffbb00; font-weight: bold;">Judging by the histogram, theoretically, there couls be about a 20% or lower chance of randomly selecting a game made in 2020 and after. The chard is very skewed to the left, and it appears there was a boom in game releases during 2010 - 2019, with about twice as many games being released in 2015 and 2016 than in 2020 and barely any in recent times.</p>

```python
df_2020_or_after = df[df['released'].dt.year >= 2020]

total_games = len(df)

games_2020_or_after = len(df_2020_or_after)

probability = games_2020_or_after / total_games

num_trials = 10000
success_count = 0

for _ in range(num_trials):
    random_game = df.sample()
    if random_game['released'].dt.year.values[0] >= 2020:
        success_count += 1

empirical_probability = success_count / num_trials

print(f"Calculated probability: {probability}")
print(f"Empirical probability after {num_trials} trials: {empirical_probability}")
```

<p style="text-align: center; color: #bb88ed; font-weight: bold;">Wow. I was wrong. The chances of getting a game released in 2020 and after are around 9.3%, which is incredibly low compared to my theoretical 20%. This could be due to the more than 100 games released every year during 2011- 2019, with some years even reaching over 175 games despite the fact that less than 100 games were released every year previously.</p>

```python
games_per_year = df['release_year'].value_counts().sort_index()

least_games_year = games_per_year.idxmin()
least_games_count = games_per_year.min()

print(f"The year with the least amount of games released is {least_games_year} with {least_games_count} games.")
```

```python
most_games_year = games_per_year.idxmax()
most_games_count = games_per_year.max()

print(f"The year with the most amount of games released is {most_games_year} with {most_games_count} games.")
```
